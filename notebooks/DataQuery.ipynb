```python
import requests
import json
import pandas as pd

email='<email-id>'
api_key='<api-key>'
```

```python
## RUN THE FOLLOWING CODE TO GET LATEST options order book DATA FOR ALL THE COINS
```

```python
asset='BTC'
yyyymmdd=20230601

# url: str = 'https://api.quantuna.com/api/fetchGSheetData'
url: str = 'http://localhost:8060/api/fetchGSheetData'
params: dict = {"product_key": "optionsOrderBookData", "coin": asset, "start_date": yyyymmdd,
                "end_date": yyyymmdd, "email": email, "api_key": api_key}
headers = {"Content-Type": "application/json; charset=utf-8"}
x = requests.post(url, json=params, headers=headers, timeout=500)
jsns = x.json()
for jsn in jsns:
    del jsn['underlyingIndexLight']

# lets get the dataframe now
df_oobd = pd.DataFrame.from_records(jsns)

# rename the data-frame columns to match to the calcDB.deribit_option_data table columns
df_oobd.columns = (df_oobd.columns.str.replace(
    '(?<=[a-z])(?=[A-Z])', '_', regex=True).str.lower())

# prepare timestamp column and return result
df_oobd['timestamp'] = pd.to_datetime(df_oobd['timestamp'], errors='coerce')
df_oobd
```
